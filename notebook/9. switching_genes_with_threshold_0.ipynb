## The switching-gene:
    1. Only on training set
    2. The genes with sequence length between 200-5000 
    3. exculde the cell line H1.hESC
    4. Two cyto-nuclear class thresholds were tested:
        aa. 0. CN-RCI >= 0 is cyto, otherwise nuclear
    5. The switching genes: 
        aa. both the numbers of cell lines for cyto and nucleus are greater than or equal to 1
        bb. the absolute value of difference between cyto and nucleus abs(cyto-nucleus) <= 1 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import openpyxl
from pandas import ExcelWriter
```

```python
def isswitch(row):
    if abs(row["Cyto-Nuclear"]) <= 1 and row["Nucleus"]>1 and row["Cyto"]>1:
        return 1
    else:
        return 0
```

```python
path_input = "../data/"
path_output = "../results/switching_genes/"
os.makedirs(path_output, exist_ok=True)
```

```python
wb = openpyxl.Workbook()
output_file = path_input + "Supplementary_Table2_185_switching_genes.xlsx"
wb.save(output_file)
writer = ExcelWriter(output_file, engine='openpyxl',mode='a')
```

```python
_df = pd.read_csv(path_input+"Supplementary_Table3_lncRNA_train_seq_with_RCI_longest_transcripts.csv",index_col=0)
```

```python
dff = _df.iloc[:,-19:].drop(["biotype","length","sequence","H1.hESC"],axis=1)

dff
```

```python
new_dff_ = dff.copy()
new_dff_ = new_dff_.set_index("gene_id")
new_dff_ = new_dff_.dropna(how="all")
```

```python
new_dff_ 
```

```python
new_dff = new_dff_.copy()
```

# threshold 0

```python
new_dff.iloc[:,0:14][new_dff.iloc[:,0:14] >= 0] = 1    # Set values >= 0 to 1
new_dff.iloc[:,0:14][new_dff.iloc[:,0:14] < 0] = 0    # Set values < 0 to 0

new_dff.iloc[:,0:14][new_dff.iloc[:,0:14].isna()] = np.nan

new_dff["Total_cell_line"] = new_dff.iloc[:,0:14].notnull().sum(axis=1)
# Add a column to store the number of columns with each value

new_dff["Nucleus"] = (new_dff.iloc[:,0:14] == 0).sum(axis=1)
new_dff["Cyto"] = (new_dff.iloc[:,0:14] == 1).sum(axis=1)
new_dff["Cyto-Nuclear"] = new_dff["Cyto"] - new_dff["Nucleus"]
new_dff["Is_switching_gene"] = new_dff.apply(lambda row:isswitch(row),axis=1)
new_dff = new_dff.reset_index()
new_dff_ = new_dff_.reset_index()
dff_switch0 = new_dff_.merge(new_dff,on="gene_id",how="left").sort_values(["Is_switching_gene"],ascending=False).reset_index(drop=True)
dff_switch0.to_excel(writer,sheet_name="switch_gene_threshhold_0")
writer.close()
```
