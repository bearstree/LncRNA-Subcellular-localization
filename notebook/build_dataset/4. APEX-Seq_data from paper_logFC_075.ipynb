```python
df_all = xl.parse("Gene lists and orphans")
df_all
```

```python
# List of log2FC columns
log2fc_columns = [
    'Nucleus_log2FC', 'Nucleolus_log2FC', 'Lamina_log2FC', 'Nuclear_Pore_log2FC',
    'Cytosol_log2FC', 'ERM_log2FC', 'OMM_log2FC', 'ER_Lumen_log2FC']

# Using threshold >=0.75 class 1, otherwise 0
log2fc_binary_df = (df_all[log2fc_columns] >= 0.75).astype(int)
log2fc_binary_df.columns = [col.replace('_log2FC', '_bin_075') for col in log2fc_binary_df.columns]

df_expanded = pd.concat([df_all, log2fc_binary_df], axis=1)
```

```python
log2fc_binary_df.columns
```

```python
nuclear = ['Nucleus_bin_075', 'Nucleolus_bin_075', 'Lamina_bin_075']
cytoplasmic = ['Cytosol_bin_075', 'ERM_bin_075', 'OMM_bin_075','ER_Lumen_bin_075']
boundry = ['Nuclear_Pore_bin_075']
```

```python
def assign_class(row):
    if (row[nuclear].sum() >= 1 ) and (row[cytoplasmic].sum()==0) and (row[boundry].sum()==0):
        return 0
    elif (row[nuclear].sum() == 0 ) and (row[cytoplasmic].sum()>=1) and (row[boundry].sum()==0):
        return 1
#     elif  (row[nuclear].sum() == 0 ) and (row[cytoplasmic].sum()==0) and (row[boundry].sum()==1):
#         return 3
    else:
        return 2
```

```python
df_expanded.columns
```

```python
df_expanded["Class"] = df_expanded.apply(lambda row:assign_class(row),axis=1) 
```
